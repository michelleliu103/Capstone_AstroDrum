```python
from scipy.io import wavfile
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd
from utils import sampleSplicing as sp
%matplotlib inline
```

```python
#path to files
filePath = '/Users/benthompson/Google Drive/MIDS/210/w210-Capstone-AstroDrum/Kick Drums/'
```

```python
#set number of iterations
#output will be n * 2 samnples
n = 10000

#iterate and create new sounds
for s in range(645,n):
    
    try:
        #get two kick filenames from path
        kick1, kick2 = sp.randFiles(filePath)
        print('iteration {}: {} and {} are the kick files being used.'.format(s,kick1,kick2))

        #read in kick files
        k1 = wavfile.read(filePath+kick1)
        k2 = wavfile.read(filePath+kick2)

        #randomly choose ms for split
        choices = [20,30,40,50,60,70,80,90,100]
        ms = np.random.choice(choices)
        print('ms is {}'.format(ms))

        #split kick 1 and 2
        k1_1,k1_2 = sp.splice(k1,ms,stereo=True)
        k2_1,k2_2 = sp.splice(k2,ms,stereo=True)

        #melt into 2 new sounds
        meltedk1 = sp.melt(k1_1,k2_2)
        meltedk2 = sp.melt(k2_1,k1_2)

        #export samples
        wavfile.write('/Users/benthompson/Google Drive/MIDS/210/w210-Capstone-AstroDrum/meltedKicks/meltedKick_{}_1.wav'.format(s),rate=44100,data=meltedk1[:,0])
        wavfile.write('/Users/benthompson/Google Drive/MIDS/210/w210-Capstone-AstroDrum/meltedKicks/meltedKick_{}_2.wav'.format(s),rate=44100,data=meltedk2[:,0])
    
    except ValueError:
        continue
```
